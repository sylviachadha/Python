# Tuple

## Tuple is a data structure that can store any data type.
## But tuples are immutable: Once a tuple is created we can not update data inside it.

## Slicing and indexing (positive and negative): Works same as that of Lists

```python
## Creating sub-tuple from a tuple 
tuple1 = (1, 2, 3, 4, 5) 

print(tuple1[1:])    # (2,3,4,5)

print(tuple1[::-2])  # It depend on step size from where u start (0 idx or -1 idx) and end
                     # So default starting value depends on step size if nothing given in start & end

print(tuple1[2:4])   # (3,4)    
```

## negative indexing also works in Tuples

```python
tuple1 = (1, 2, 3, 4, 5)   ## -1 = 5, -2 = 4, -3 = 3

print(tuple1[-1])     # 5 

print(tuple1[-4])     # 2 

print(tuple1[-3:-1])  # 3,4 

print(tuple1[:-1])    # ()   ???

print(tuple1[-2:])    # 4,5 since step size +ve, move from left to right

print(tuple1[0:1])    # 1,
print("Notice the last print statement. A single element of a tuple is always represented with a comma sign after it")
```

## Representation: Packing and unpacking of Tuples

#### Tuple packing refers to assigning multiple values into a tuple. Tuple unpacking refers to assigning a tuple into multiple variables

### Changing a Tuple

    Unlike lists, tuples are immutable.
    This means that elements of a tuple cannot be changed once they have been assigned. 
    But, if the element is itself a mutable data type like a list, its nested items can be changed.
    We can also assign a tuple to different values (reassignment).

```python
# Changing tuple values
my_tuple = (4, 2, 3, [6, 5])
print(my_tuple)  # ((4, 2, 3, [6, 5]))

# TypeError: 'tuple' object does not support item assignment
my_tuple[1] = 9

## Individual elements of tuple cannot be changed.
```

```python
# However, item of mutable element can be changed
my_tuple[3][0] = 9    # Output: (4, 2, 3, [9, 5])
print(my_tuple)

# QUESTION
# Tuples can be reassigned - it means only whole tuple can be reassigned/overwritten but 
                            #individual elements can't update since immutable type
my_tuple = ('p', 'r', 'o', 'g', 'r', 'a', 'm', 'i', 'z')

# Output: ('p', 'r', 'o', 'g', 'r', 'a', 'm', 'i', 'z')
print(my_tuple)
```

```python
# list inside tuples
favorites = ('Chennai', ['Delhi'], 'Bangluru')
print(len(favorites))
print(type(favorites))
##favorites[1].pop() # at postion 1 we have list so we can use pop.

favorites[1].append("Haridwar")
print(favorites)
print(len(favorites))

favorites[1].pop()
print(favorites)
###print(sorted(favorites))

print(type(favorites))
```

## List vs. Tuple

#### Advantages of Tuple over List
Since tuples are quite similar to lists, both of them can be used in similar situations. However, there are certain advantages of implementing a tuple over a list. Below listed are some of the main advantages:

    We generally use tuples for heterogeneous (different) data types and lists for homogeneous (similar) data types.
    Since tuples are immutable, iterating through a tuple is faster than with list. So there is a slight performance  boost.
    Tuples that contain immutable elements can be used as a key for a dictionary. With lists, this is not possible.
    If you have data that doesn't change, implementing it as tuple will guarantee that it remains write-protected.
