```python
# Slicing in Python

# Slicing the list basically means taking the subset out of your list
# Slicing in Python is equivalent to subsetting in R
```

```python
list1 = ["A","B","C","D","E","F","G","H","I","J"]
```

```python
list1
```

```python
list1[:] # Full big slice
```

```python
list1[2:] # Start from 2 and go till end 
```

```python
list1[:7] # Start at beginning and go only until 7 which will mean index 0-6 and total
# 7 elements 
```

```python
# 7 is not inclusive
slice_list1 = list1[2:7]
```

```python
slice_list1
```

```python
# Negative indexation in slicing
slice_list2 = list1[-8:-3]
```

```python
slice_list2
```

```python
slice_list3 = list1[-8:7]
slice_list3
```

```python
slice_list4 = list1[2:-3]
slice_list4
```

```python
# Advanced Slicing including start, end and step size
# Similar to range function
# Below will return elements 2,4,6,8 as 9 is not inclusive, whichever is last is not 
# exclusive

slice_list5 = list1[2:9:2]
slice_list5
```
