```python
# Access elements of a list using indexation
# There are 5 elements however indexation is from 0 to 4

l1 = ['a', 'b', 'c', 'd', 'e']
l1
```

```python
l1[0]
```

```python
l1[4]
```

```python
# Normal Indexation and Negative indexation in R to access elements
# 0  1  2  3  4  
#-5 -4 -3 -2 -1
```

```python
l1[-1] # In R this would mean you need all elements of vector except the
       # the element with index -1 that will be first element in R as 
       # indexation in R starts from 1 and not 0 as in case of Python
       # However in python this means negative indexation to access the
       # elements rather than exclusion as in R
```

```python
l1[-5]
```

```python
l1[2]  # l1[2] and l1[-3] are same
```

```python
l1[-3]
```

```python
# If there are many elements in list and you would like to access last 
# element then to know no of elements use len function (gives length)
# For eg here length is 5 means 5 elements in list however indexation
# starts from 0 so last element will be accessed by l1[4]
```

```python
len(l1)
```
