### Precedence Operators in Python that affect order of operations in Python
#### The operator precedence in Python is listed in the following table. It is in descending order (upper group has higher precedence than the lower ones).

### When two operators have the same precedence, associativity helps to determine the order of operations.
##### Associativity is the order in which an expression is evaluated that has multiple operators of the same precedence. Almost all the operators have left-to-right associativity. For example, multiplication and floor division have the same precedence. Hence, if both of them are present in an expression, the left one is evaluated first.



```python
print((2+3)*5/2%6)     # precedence rule again
print(2**3**2)        # exponents associativity right to left ***
```

## Bitwise Operators:
* In Python, bitwise operators are used to performing bitwise calculations on integers. 
* The integers are first converted into binary and then operations are performed on bit by bit. 
**Finally the result is returned in decimal format.**

## Logic Gates



### AND (&) operator returns 1 if both the bits are 1 else 0

### Example a = 10 and b = 4
* a= 10 = 1010 (in binary)
* b = 4 = 0100 (in binary)
* 0000 (in case of & operator)
* 1110 (in case of | operator)


```python
#Example for a = 10 and b = 4
a = 10
b = 4
print(a & b)
```

### Example a = 10 and b = 4 (a | b)
* a= 10 = 1010 (in binary)
* b = 4 = 0100 (in binary)
* 1110 (in case of | operator)

### OR (|) operator returns 1 if either of the bit is 1 else 0

```python
print(a | b)
```

### XOR (^) operator returns 1 only if one of the bits is 1 and the other is 0 else returns false.

```python
print(a ^ b)
```

### not (~) operator returns ones complement of the number.
    ones complement means adding one to the binary of the given integer
    a = 10 = 1010 (in binary)

```python
print (~a)
```

### Shift Operators
#### They can be used when we have to multiply (left shift) or divide (right shift) a number by two
#### Generally ... 2^32,  2^16, 2^8, 2^4, 2^2, 2^1, 2^0
#### So when shifting to right, no. is getting divided by 2 & when u shift to left no. is getting multiplied by 2

### Right Shift: Bitwise right shift: Shifts the bits of the number to the right and fills 0 on voids left
#####  8 >> n is equivalent to 8 // 2**n

```python
a = 10 ### 0000 1010 (Binary)
a >> 1 ### 0000 0101 = 5
```

### Bitwise left shift: Shifts the bits of the number to the left and fills 0 on voids right as a result. 

```python
b=10  ### 0000 1010
b << 1 ### 0001 0100
```
