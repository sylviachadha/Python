## Data visualization using Matplotlib
### Matplotlib is strongly object-oriented & its principal objects are figure and axes.

1. Visualization is important. Data in table not very interpretable e.g. 1 million rows in table, cannot see trends with table/df. U can use charts to visualize. A large dataset can be compressed into 1 single visual depicting more information. Common charts include line chart, bar chart, histogram, pie chart.
2. Library - matplotlib, this plotting library is based on syntax of tool called matlab.
3. MATLAB is a proprietary multi-paradigm programming language and numeric computing environment (allows matrix multiplications) developed by MathWorks.
4. Any chart u create in matplotlib u can also export in pdf format or image format etc


```python
# Import matplotlib's pyplot
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

Understand figure and axes components
1. using explicit method/object-oriented - we create the object-figure,axes & then charts
2. using implicit method

### Explicit method

```python
# Method 1 - explicit method

fig, ax = plt.subplots()   # by default 1 figure & 1 axis
# u can also have common x/y axis for multiple plots using subplots
```

```python
fig, ax = plt.subplots(2,3)
```

```python
x = np.linspace(0,2,100)
print(x)
```

```python
fig, ax = plt.subplots(figsize=(10,5))      # on same axes, u r plotting 3 lines.
ax.plot(x,x,label='linear')     # linear line
ax.plot(x,x**2,label='quadratic')  # quadratic line
ax.plot(x,x**3, label='cubic') # cubic line
ax.set_xlabel('x-label')
ax.set_ylabel('y-label')
ax.set_title('Simple PLot')
ax.legend();
```

### Implicit Method

```python
# Method 2 - Implicit method (it implicitly assumes a fig and axes) u just need to use (plt) inbuilt function.

plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple PLot')
plt.legend();
```

```python
# Here since we do not have fig,ax we cannot control figure size.
# In case we want to adjust size, we explicitly need to create a figure here as well.

plt.figure(figsize=(10,5))
plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple PLot')
plt.legend();
```

Note -
For simple charts or quick exploration, better to use implicit charts.
For complex charts & to gain better flexibility over settings & features of charts, use explicit method.
