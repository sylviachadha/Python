```python
# HW_Core_Programming
# Prove law of large numbers
# Change n from 100 , 1000, 10000, 100000 and average expected should come close to
# 68.2%. As u increase n mean comes to expected value of 68.2% which proves law of 
# large numbers

# Function to generate random number is randn
import numpy as np
from numpy.random import randn
counter = 0
n = 100000

# to generate 100 random value
# randn(100)
# Iterate over 100 random values so instead of range we use randn(n)
for i in randn(n):
    if (i < 1 and i > -1):
        counter = counter + 1
print (counter)
average = counter / n
print(average)
                
```
