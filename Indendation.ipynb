```python
# Indendation is a very important part of Python and how code can go
# wrong if indendation is not correct

# Indendation - Code not in Else statment

import numpy as np
from numpy.random import randn

x = randn()
answer = None
if x > 1:
    answer = "Greater than 1"
elif x >= -1:
    answer = "Between -1 and 1"
else:
    answer = "Less than -1"
print ("Hello")
print (x)
print (answer)
```
