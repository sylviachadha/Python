### Question 8

Write a program to slice a given list into 3 equal chunks and reverse each chunk.





### Solution 8

```python
list1 = [2,4,6,8,10,12,14,16,18]

chunked_list=[]
reverse_list=[]
chunk_size=3

for i in range(0, len(list1), chunk_size):
    chunk = list1[i:i+chunk_size]
    chunked_list.append(chunk)
    rev_chunk = chunk[-1::-1]
    reverse_list.append(rev_chunk)
  

print('\nChunk1',chunked_list[0])
print('After reversing it',reverse_list[0])
     
print('\nChunk2',chunked_list[1])
print('After reversing it',reverse_list[1])

print('\nChunk3',chunked_list[2])
print('After reversing it',reverse_list[2])
```

### Question 10

Write a program to sort a list alphabetically in a dictionary



### Solution 10

```python
my_dict = {'n1':['a','A','c'],
          'n2':['9','1','w'],
          'n3':['5','9','0']}

print('Original\n', my_dict,'\n')

sorted_dict={}
for key,value in my_dict.items():
    sorted_val = sorted(value)
    sorted_dict[key]=sorted_val

print('Sorted\n', sorted_dict)
```

### Question 11

Write a program to calculate the average value of the numbers in a given tuple of tuples.



### Solution 11

```python
# * give elements of nested tuple- Tuple1[0], Tuple1[1], Tuple1[2],every element is iterator
# zip() function creates an iterator that will aggregate elements from two or more iterables.


Tuple1 = ((10,10,20,10),(12,14,15,25),(16,18,13,17))
print('\n Tuple1=',Tuple1,'\n')

list1=[]
for i in zip(*Tuple1): 
    print(i)
    result = sum(i) / len(i)
    list1.append(result)

Tuple2 = tuple(list1)
print('\n Tuple2=',Tuple2)
```

### Question 12

Write a program to convert a given tuple of positive integers into an integer.



### Solution 12

```python
# The map(func, iterable) function executes a specified function for each item in an iterable.
# The str() function converts the specified value into a string.
# ''.join will join elements of tuple with specified seperator like '',# etc


def tuple_to_int(Tuple):
    result = int(''.join(map(str,Tuple)))
    return result

Tuple = (1,2,3)
print("Input: ") 
print(Tuple)

print("Output:")
print(tuple_to_int(Tuple))
```
