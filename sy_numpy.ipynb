## Linear vs Lograithmic Scale

The key difference between linear and logarithmic scales lies in how the y-axis represents the values.

The x-axis does not make any difference, it is used to indicate the position of each point, there are total 10 points.

```python
import numpy as np
import matplotlib.pyplot as plt

# Logarithmically spaced values
log_values = np.array([1, 1.66810054, 2.7825594, 4.64158883, 7.74263683, 12.91549665, 
                       21.5443469, 35.93813664, 59.94842503, 100])

# Indices for the x-axis
indices = np.arange(1, 11)

# Plot on a linear scale
plt.figure(figsize=(14, 6))

# Linear Scale Plot
plt.subplot(1, 2, 1)
plt.plot(indices, log_values, 'o-', label='Logarithmic Values (Linear Scale)')
plt.title('Logarithmic Values on Linear Scale')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
for i, txt in enumerate(log_values):
    plt.annotate(f'{txt:.2f}', (indices[i], log_values[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Logarithmic Scale Plot
plt.subplot(1, 2, 2)
plt.plot(indices, log_values, 'o-', label='Logarithmic Values (Log Scale)')
plt.title('Logarithmic Values on Logarithmic Scale')
plt.xlabel('Index')
plt.ylabel('Value')
plt.yscale('log')
plt.legend()
for i, txt in enumerate(log_values):
    plt.annotate(f'{txt:.2f}', (indices[i], log_values[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()
```

The term "exponential scale" can be somewhat ambiguous because there isn't a standard "exponential scale" for axes in plotting, but we can illustrate how exponential growth looks when using a linear scale and a logarithmic scale.

```python
import numpy as np
import matplotlib.pyplot as plt

# Exponential growth values
base = 2
a = 1  # Initial value
indices = np.arange(10)
exp_values = a * (base ** indices)

# Linear scale plot
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(indices, exp_values, 'o-', label='Exponential Growth (Linear Scale)')
plt.title('Exponential Growth on Linear Scale')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
for i, txt in enumerate(exp_values):
    plt.annotate(f'{txt:.2f}', (indices[i], exp_values[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Logarithmic scale plot
plt.subplot(1, 2, 2)
plt.plot(indices, exp_values, 'o-', label='Exponential Growth (Log Scale)')
plt.title('Exponential Growth on Logarithmic Scale')
plt.xlabel('Index')
plt.ylabel('Value')
plt.yscale('log')
plt.legend()
for i, txt in enumerate(exp_values):
    plt.annotate(f'{txt:.2f}', (indices[i], exp_values[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()
```

Exponential growth appears as a straight line on a logarithmic scale, making it easier to interpret trends and compare rates of change. Logarithmic scales can compress large ranges of data (cz it is like 1,10,100 etc), making it easier to visualize and analyze data that spans several orders of magnitude. Logarithmic scales highlight relative changes rather than absolute changes, which is often more meaningful in exponential data.

## Absolute vs Relative Scale

- Absolute change is useful for understanding the exact amount of change in a value, but it doesn't provide context regarding the scale or proportion of the change.

- Relative change is the ratio of the absolute change to the initial value. It measures the percentage increase or decrease relative to the initial value.

```python
import numpy as np
import matplotlib.pyplot as plt

# Exponential growth values (powers of 10)
values = np.array([10**0, 10**1, 10**2, 10**3])

# Indices for the x-axis
indices = np.arange(len(values))

# Plot on a linear scale
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(indices, values, 'o-', label='Exponential Growth (Linear Scale)')
plt.title('Exponential Growth on Linear Scale')
plt.xlabel('Index')
plt.ylabel('Value')
plt.yscale('linear')
plt.legend()
for i, txt in enumerate(values):
    plt.annotate(f'{txt}', (indices[i], values[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Plot on a logarithmic scale
plt.subplot(1, 2, 2)
plt.plot(indices, values, 'o-', label='Exponential Growth (Log Scale)')
plt.title('Exponential Growth on Logarithmic Scale')
plt.xlabel('Index')
plt.ylabel('Value')
plt.yscale('log')
plt.legend()
for i, txt in enumerate(values):
    plt.annotate(f'{txt}', (indices[i], values[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()
```
