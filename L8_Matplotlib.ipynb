## Difference b/w Bar chart and Histogram

Histograms plot quantitative data/numerical data with ranges of the data grouped into bins or intervals while bar charts plot categorical data.




### Histogram

```python
fig, ((ax1), (ax2))= plt.subplots(nrows=1,ncols=2,figsize=(10,5))  # (width,height)

#x[idx].hist(df.iloc[:,idx], bins=12, alpha=0.5)

ax1.hist(sample['Salary'], bins='auto', alpha = 0.5)
ax2.hist(sample['Age'], bins= 'auto', alpha = 0.3)
ax1.set_title("Salary histogram")
ax1.set_xlabel("Salary")
ax1.set_ylabel("freq");
ax2.set_title("Age histogram")
ax2.set_xlabel("Age")
ax2.set_ylabel("freq");
```

### Overlapping the plots

```python
'''Overlapping Plots'''
import numpy as np

# Earlier we used data in form of pandas dataframe (df.plot) now we r using data in form of 
# array data structure. Mostly in case of data structure numpy array, the plot 
# command (using matplotlib.pyplot as plt.plot) will by default plot line chart.
# plt.scatter for scatter plot.

x1 = np.array([0, 1, 2, 3])
y1 = np.array([2, 4, 7, 9])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([1, 3, 9, 11])
x3 = np.array([0, 1, 2, 3])
y3 = np.array([5, 8, 3, 10])

# In pandas with dataframe when u use pandas plot function u have to exclusively specify what
# is x=,y=,kind=bar/line/scatter but with matplotlib plot function with array data structure
# by default it plots line
 
plt.plot(x1, y1, x2, y2) #plot generated line graph by default in case of normal data structure
plt.scatter(x3,y3)
plt.title("Title of the plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis");

# In matplotlib we need to exclusively define axis but in pandas plot since we pass x='',y'' 
# in plot function so it takes axes from there only inbuilt feature, no need to specify 
# explicitly.
```

#### Bar Plot without annotations

```python
'''x-coordinate: bar.get_x() + bar.get_width() / 2 
y-coordinate: bar.get_height() 
free space to be left to make graph pleasing: (0, 8) 
ha and va stand for the horizontal and vertical alignment'''

food = np.array([10, 25, 20, 15, 22])
put_labels = ["MILK", "BUTTER", "BREAD", "TEA", 'COFFEE']

x = np.arange(len(put_labels))  # the label locations
print('x-tick positions',x)
#width = 0.5  # the width of the bars 

plots = plt.bar(x,food)
plt.xticks(ticks = x, labels = put_labels); #xticks takes(tick locations,tick labels)
```

####  Bar Plot with annotations

```python
'''x-coordinate: bar.get_x() + bar.get_width() / 2 
y-coordinate: bar.get_height() 
free space to be left to make graph pleasing: (0, 8) 
ha and va stand for the horizontal and vertical alignment'''

food = np.array([10, 25, 20, 15, 22])
put_labels = ["MILK", "BUTTER", "BREAD", "TEA", 'COFFEE']

x = np.arange(len(put_labels))  # the label locations
print('x-tick positions',x)
#width = 0.5  # the width of the bars 

plots = plt.bar(x,food)
plt.xticks(ticks = x, labels = put_labels) #xticks takes(tick locations,tick labels)

#-----------------------------------------------------------------------#
## ADDITIONAL CODE FOR ADDING ANNOTATIONS ONLY USING annotate function ##

#iterate over every bar
for bar in plots:
    plt.annotate(format(bar.get_height(), '.2f'),  # .2f to get upto 2 decimal places
                   (bar.get_x() + bar.get_width() / 2,  
                    bar.get_height()), ha='center', va='center', 
                   size=10, xytext=(0,5), 
                   textcoords='offset points')
    
#plt.annotate(text, xy) - Annotate the point *xy* with text *text*
```

### Plot with Spline

```python
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()
x = np.linspace(-np.pi, np.pi, 100) # pie used to represent trignometric functions like sin,cos
y = 2*np.sin(x)

ax = fig.add_subplot(111)  

# These are subplot grid parameters encoded as a single integer. For example, "111" means
# "1x1 grid, first subplot" and "234" means "2x3 grid, 4th subplot".
# Special plots to represent some mathematical o/p.

# Spines to divide the graph and bring it to the center (0)
ax.set_title('zeroed spines')
ax.plot(x, y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# remove the ticks from the top and right edges
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a


x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,sig)
plt.show()

# Splines are generally set at center because usually behaviour of curve changes at center 
# & we want to see that change like after 0 changing to +ve from -ve.
```
