```python
a = np.array(l)
```

```python
type(a)
```

```python
a
```

```python
a[2:]
```

```python
a[2:4]
```

```python
# Whenever u slice an array u do not create a copy of that array
# while with list u r creating another copy so u can work with it 
# seperately.
# U just create a view of an array when working with arrays so if u
# make changes to the slice u r going to make changes to the original
# array
```

```python
b = a[2:4]
```

```python
print(b)
```

```python
#Here u have not created another array b but u have created a view of
# array a 
```

```python
b[0]
```

```python
b[:]
```

```python
b[:] = 111
```

```python
b
```

```python
a
```

```python
# if u change something in b the value in original a array also have 
# been changed as shown above 
# So when u r creating slice of an array u r not creating a new array
# u r actually just working with same old array just creating a view
# of that array
# That's just how python/numpy protects memory of computer
```

```python
# In case we need to create a copy of an array use copy function as below
c = a.copy()
```

```python
c
```

```python
# Now we make changes to c
c[1] = 9
```

```python
c
```

```python
a
```

```python
# So as seen above when we create a copy a remains as it is and changes
# in c remain seperately in c and do not affect original array a
```
