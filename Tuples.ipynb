```python
type (letters)
```

```python
letters[::-1]
```

```python
letters[2:7:1]
```

```python
letters[2:7:-1]  # Here 7 is not inclusive 
```

```python
# Nothing appears above as starting from 2 we can never reach 7 backwards so we
# need to change order from 7 to 2 if we have to give a negative step size 
```

```python
letters[7:2:-1] 
```

```python
letters[6:1:-1]  # here 1 is not inclusive
```
