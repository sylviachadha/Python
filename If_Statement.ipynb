```python
# If , elif and else statement in Pyhton
# If statement
# Generate random numbers that are normally distributed that is (mean of
# 0 and standard deviation of 1) and check where they fall

# numpy is a package for scientific computation / calculation
import numpy as np
from numpy.random import randn

```

```python
randn()
```

```python
#---- -2 ---- -1 ---- 0 ---- 1 ---- 2 ----
# In python None is like Null in other programming languages
# None used to initialize a variable
# Difference b/w if and while is that if does not come back and recheck
# the logical condition while while does
x = randn()
answer = None
if x > 1:
    answer = "Greater than 1"
print (x)
print (answer)
```

---

```python
# Else statement
x = randn()
answer = None
if x > 1:
    answer = "Greater than 1"
else:
    answer = "less than 1"
print (x)
print (answer)
```

```python
#---- -2 ---- -1 ---- 0 ---- 1 ---- 2 ----
# Nested statements
x = randn()
answer = None
if x > 1:
    answer = "Greater than 1"
else:
    if x >= -1:
        answer = "Between -1 and 1"
    else:
        answer = "Less than -1"
print(x)
print(answer)
    
```

```python
# else if in python called as elif
x = randn()
answer = None
if x > 1:
    answer = "Greater than 1"
elif x >= -1:
    answer = "Between -1 and 1"
# In else which is last condition when none of above is true you do not
# need to put any condition explicitly as it means when no other 
# condition satisfied execute else
else:
    answer = "Less than -1"
print(x)
print(answer)
      
    
```
